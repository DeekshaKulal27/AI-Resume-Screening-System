# Import Libraries

```python
from langchain_core.prompts import PromptTemplate
from langchain_community.llms import Ollama
```

# Connect to Ollama

```python
llm = Ollama(model="llama3")
```

# Test Model

```python
print(llm.invoke("Explain AI in one line"))
```

# Job Description

```python
job_description = """
Job Role: Data Scientist

Required Skills:
- Python
- Machine Learning
- Deep Learning
- SQL
- Data Visualization

Experience Required:
- Minimum 2+ years

Tools:
- TensorFlow
- Pandas
- Scikit-learn
"""
```

# Resumes

```python
strong_resume = """
Skills: Python, Machine Learning, Deep Learning, SQL
Tools: TensorFlow
Experience: 3 years
"""

average_resume = """
Skills: Python, SQL, Basic Machine Learning
Experience: 1 year
"""

weak_resume = """
Skills: HTML, CSS, Excel
Experience: No relevant experience
"""
```

# Prompt 1 – Extraction

```python
extract_prompt = PromptTemplate(
    input_variables=["resume"],
    template="""
Extract the following:

Skills:
Tools:
Experience:

Resume:
{resume}
"""
)
```

# Function – Extraction

```python
def extract_skills(resume):
    prompt = extract_prompt.format(resume=resume)
    return llm.invoke(prompt)
```

# Prompt 2 – Matching

```python
match_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Compare resume with job description.

Return STRICTLY:

Matching Skills: [skills]
Missing Skills: [skills]

Resume:
{resume}

Job:
{job}
"""
)
```

# Function – Matching

```python
def match_skills(resume):
    prompt = match_prompt.format(resume=resume, job=job_description)
    return llm.invoke(prompt)
```

 # Prompt 3 – Scoring

```python
score_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Give score (0–100) based on match.

Return:
Score: <number>

Resume:
{resume}

Job:
{job}
"""
)
```

# 12: Function – Scoring

```python
def score_resume(resume):
    prompt = score_prompt.format(resume=resume, job=job_description)
    return llm.invoke(prompt)
```

# Prompt 4 – Explanation

```python
explain_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Explain evaluation.

Return:

Strengths:
- ...

Weaknesses:
- ...

Final Justification:
- ...

Resume:
{resume}

Job:
{job}
"""
)
```

# Function – Explanation

```python
def explain_resume(resume):
    prompt = explain_prompt.format(resume=resume, job=job_description)
    return llm.invoke(prompt)
```

# Pipeline Function

```python
def run_pipeline(resume):
    print("===== EXTRACTION =====")
    print(extract_skills(resume))

    print("\n===== MATCHING =====")
    print(match_skills(resume))

    print("\n===== SCORE =====")
    print(score_resume(resume))

    print("\n===== EXPLANATION =====")
    print(explain_resume(resume))
```

# Run for Candidates

```python
print("🔷 STRONG CANDIDATE")
run_pipeline(strong_resume)

print("\n🔷 AVERAGE CANDIDATE")
run_pipeline(average_resume)

print("\n🔷 WEAK CANDIDATE")
run_pipeline(weak_resume)
```

Debug Example 

```python
wrong_resume = "I am expert in everything"
run_pipeline(wrong_resume)
```
